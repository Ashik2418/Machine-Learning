```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer,SimpleImputer
```

```python
df = pd.read_csv('/content/Titanic wala dataset.csv')[['Age','Pclass','Fare','Survived']]
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=['Survived'],axis=1)
y = df['Survived']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

# **Using normal SimpleImputer --> mean**

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_trf,y_train)
y_pred = lr.predict(X_test_trf)
accuracy_score(y_test,y_pred)
```

# **Now Using KNN Imputer**

```python
knn = KNNImputer(n_neighbors=5,weights='distance')
X_train_trf2 = knn.fit_transform(X_train)
X_test_trf2 = knn.transform(X_test)
```

```python
pd.DataFrame(X_train_trf2,columns=X_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(X_train_trf2,y_train)
y_pred2 = lr.predict(X_test_trf2)
accuracy_score(y_test,y_pred2)
```
