# **Working with Dates**

```python
# Converting to datetime datatype
date['Delivery Date'] = pd.to_datetime(date['Delivery Date'])
date.info()
```

# *1.Extract year**

```python
date['date_year'] = date['Delivery Date'].dt.year
date.head()
```

# *2.Extract Month*

```python
date['date_month_no'] = date['Delivery Date'].dt.month
date.head()
```

# *3.Extract date_month_name*

```python
date['date_month_name'] = date['Delivery Date'].dt.month_name()
date.sample(5)
```

# *Extract Days*

```python
date['date_day'] = date['Delivery Date'].dt.day
date.head()
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).head()
```

# *Extract Day of week*

```python
date['date_dow'] = date['Delivery Date'].dt.dayofweek
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
```

# *Day of week_name*

```python
date['date_dow_name'] = date['Delivery Date'].dt.day_name()
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
```

# *is this day is weekend?*

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
```

# *Extract week of the year*

```python
date['date_week'] = date['Delivery Date'].dt.isocalendar().week

date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
```

# *Extract Quarter*

```python
date['quarter'] = date['Delivery Date'].dt.quarter
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
```

# *Extract Semester*

```python
date['semester']=np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['Order ID','Customer ID','Product ID','Customer Status','Date Order was placed','Quantity Ordered','Total Retail Price for This Order','Cost Price Per Unit']).sample(5)
#
```

# **Extract Time elapsed between dates**

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['Delivery Date']
```

```python
(today - date['Delivery Date']).dt.days
```

# *Months Passed*

```python
date['months_diff'] = (today.year - date['Delivery Date'].dt.year) * 12 + (today.month - date['Delivery Date'].dt.month)
date.sample(5)
```
